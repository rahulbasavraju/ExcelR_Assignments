## 1. Airlines Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
airlines = pd.read_csv("/Users/rahul/Desktop/excelr/Data Science Assignments/Clustering/EastWestAirlines.csv")
airlines
```

```python
airlines.info()
```

```python
airlines.shape
```

```python
airlines.isna().sum() 
```

```python
airlines.describe()
```

### Hierarchy Clustering

```python
# create dendrogram
plt.figure(figsize = (16, 10));
dendrogram = sch.dendrogram(sch.linkage(airlines, method='average'))
```

```python
# Agglomerative Clustering
hc = AgglomerativeClustering(n_clusters=5, affinity = 'euclidean', linkage = 'average')
```

```python
# save clusters for chart
y_hc = hc.fit_predict(airlines)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
airlines1 = pd.concat([airlines,Clusters], axis=1)
airlines1
```

```python
airlines1.sort_values("Clusters")
```

### K-Means

```python
kmeans = KMeans(n_clusters=14,random_state=0)
kmeans.fit(airlines)
```

```python
kmeans.inertia_
```

```python
wcss = []
for i in range(1, 19):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(airlines)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 19), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## 2. Crime Dataset 

```python
c = pd.read_csv("/Users/rahul/Downloads/Data Science Assignments/Clustering/crime_data.csv")
c.head()
```

```python
crime=c.rename({'Unnamed: 0':'States'},axis=1)
crime.tail()
```

```python
crime.States.value_counts()
```

```python
crime.info()
```

### Hierarchy Clustering

```python
n = MinMaxScaler()
data= n.fit_transform(crime.iloc[:,1:].to_numpy())
crimes = pd.DataFrame(data, columns = crime.columns[1:])
crimes.head(4)
```

### Clustering

```python
# create dendrogram
fig = plt.figure(figsize=(15,8))
dendrogram = sch.dendrogram(sch.linkage(crimes, method='average'))
```

```python
# create dendrogram
fig = plt.figure(figsize=(15,8))
dendrogram = sch.dendrogram(sch.linkage(crimes, method='ward'))
```

```python
# create clusters
hc1 = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'average')
```

```python
# save clusters for chart
y2 = hc1.fit_predict(crimes)
cc = pd.DataFrame(y2,columns=['Clusters'])
cc.head(6)
```

```python
crimes2 = pd.concat([crime,cc],axis=1)
crimes2.head(7)
```

```python
crimes2.sort_values("Clusters").reset_index()
```

```python
crimes2['Clusters'].value_counts()
```

### K-Means

```python
kmeans = KMeans(n_clusters=4,random_state=0)
kmeans.fit(crimes)
```

```python
kmeans.inertia_
```

```python
wcss = []
for i in range(1, 8):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(crimes)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 8), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
cc2 = KMeans(4, random_state=8)
cc2.fit(crimes)
```

```python
cc2.labels_
```

```python
# Converting array to dataframe
df2 = pd.DataFrame(cc2.labels_, columns =['clusters'])
```

```python
crimes3 = pd.concat([crime,df2], axis=1)
crimes3.head(4)
```

```python
crimes3['clusters'].value_counts()
```

```python
crimes3.groupby('clusters').agg(['mean'])
```
